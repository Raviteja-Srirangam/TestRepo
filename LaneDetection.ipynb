```python
import numpy as np
import cv2
import math



def RegionOfInterest(img, vertices):
    mask = np.zeros_like(img)
    cv2.fillPoly(mask, vertices, 255)
    masked_image = cv2.bitwise_and(img,mask)
    return masked_image

def drawLines(img, lines, color = [0,0,255], thickness=5):
    line_img = np.copy(img)*0
    img = np.copy(img)
    if lines is not None:
        for line in lines:
            for x1,y1,x2,y2 in line:
                cv2.line(line_img,(int(x1),int(y1)),(int(x2),int(y2)),color,thickness)
    imgWithLines = cv2.addWeighted(img,0.8,line_img,1,0)
    return imgWithLines



def coordinates(image,lines):
    max_y = int(image.shape[0])
    min_y = int(max_y*3/5)
    slope,intercept = lines
    max_x = int((max_y-intercept)/slope)
    min_x = int((min_y-intercept)/slope)
    return np.array([max_x,max_y,min_x,min_y])

def getSlope(image,lines):
    left=[]
    right=[]
    global global_left_fit_average
    global global_right_fit_average
    for line in lines:
        x1,y1,x2,y2 = line.reshape(4)
        params = np.polyfit((x1,x2),(y1,y2),1)
        slope = params[0]
        intercept = params[1]
        if slope<=0:
            left.append([slope,intercept])
        else:
            right.append([slope,intercept])
    if (len(left)!=0):
        left_avg = np.average(left,axis=0)
        global_left_fit_average=left_avg
    else:
        left_avg = global_left_fit_average
    if (len(right)!=0):
        right_avg = np.average(right,axis=0)
        global_right_fit_average=right_avg
    else:
        right_avg = global_right_fit_average
    left_line = coordinates(image,left_avg)
    right_line = coordinates(image,right_avg)
    return np.array([[left_line,right_line]])


def pipeline(image):
    #image = cv2.imread("CircularWhite.png")
    height = image.shape[0]
    width = image.shape[1]
    img_mono = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    img_blur = cv2.GaussianBlur(image, (7,7),0)
    img_canny = cv2.Canny(img_blur, 20,100)
    vertices = np.array([[(0,height),(width/2,height/2),(width,height)]],dtype=np.int32)
    roiImage = RegionOfInterest(img_canny,vertices)
    lines = cv2.HoughLinesP(roiImage,1,np.pi/60, 40, np.array([]), minLineLength =40, maxLineGap = 25)
    
    
    
    
    line_array = getSlope(image,lines)
    final_lines = np.array(line_array)
    #print(final_lines)
    
    finalImage = drawLines(image, final_lines)
    return finalImage
   
        
    #cv2.imshow('Lanes',finalImage)
    #cv2.waitKey(0)
    #cv2.destroyAllWindows()
    
    
    


#pipeline("CircularWhite.png")
```
